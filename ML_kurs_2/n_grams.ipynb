```python
import nltk
import re

from nltk.lm import MLE
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.tokenize import word_tokenize, sent_tokenize 

nltk.download('punkt')
nltk.download('punkt_tab')
```

## hämta datan


```python
with open('../data collections/prince.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

## förbehandla datan
(dela in textfilen i en lista med varje ord)

```python
text= text.lower()

text= re.sub(r'[^a-zA-Z\s\.]', '', text)
sentences= sent_tokenize(text)

print(sentences[:5])
```

```python
tokenized_sentences = []

for sentence in sentences:
    words = word_tokenize(sentence)
    tokenized_sentences.append(words)

print(tokenized_sentences[:5])
```

## Skapa bigram modell

```python
n = 3

train_data, vocab = padded_everygram_pipeline(n, tokenized_sentences)

bigram_model = MLE(n)

bigram_model.fit(train_data, vocab)
```

```python
bigram_model.score('prince', ['the', 'happy']) # hur sannolikt är det att prince kommer efter happy
```

```python
generated_words= bigram_model.generate(20)
print(generated_words)
```

```python
sentence = " ".join(generated_words).replace('</s>', '')

print(sentence)
```

```python
def clean_generated_text(words):
    cleaned_words = []

    for word in words:
        if word != "<s>" and word != "</s>":
            cleaned_words.append(word)

    return " ".join(cleaned_words)
```

```python
bigram_words = bigram_model.generate(
    20,
    text_seed=["the", 'happy'],
    
)

print(clean_generated_text(bigram_words))
```
